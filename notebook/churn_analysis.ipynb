# DATA PREPROCESSING
    - Encoding
    - Train-Test Split
    - Feature Scaling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df_encoded.drop('Churn_Yes', axis=1)
y = df_encoded['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
#  Feature Scaling
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
X_test_scaled.shape
```

# TRAIN THREE MACHINE LEARNING MODELS
- Logistic Regression
- Random Forest
- XGBoost


```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train_scaled, y_train)
lr_pred = lr_model.predict(X_test_scaled)
```

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_scaled, y_train)
rf_pred = rf_model.predict(X_test_scaled)
```

```python
xgb_model = XGBClassifier(
 eval_metric='logloss',
 random_state=42
)
xgb_model.fit(X_train_scaled, y_train)
xgb_pred = xgb_model.predict(X_test_scaled)
```

```python
# ========================================
# Compare Accuracy, Precision, Recall
# and Confusion Matrix for All Models
# ==========================================

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    confusion_matrix
)
import pandas as pd

# Store all models and their corresponding test data
# Logistic Regression uses scaled data
# Random Forest and XGBoost use original (unscaled) data
models = {
    "Logistic Regression": (lr_model, X_test_scaled),
    "Random Forest": (rf_model, X_test),
    "XGBoost": (xgb_model, X_test)
}

# Empty list to store results
results = []

# Loop through each model
for model_name, (model, X_eval) in models.items():

    # Make predictions on the test set
    y_pred = model.predict(X_eval)

    # ------------------------------------------
    # 1. Accuracy
    # ------------------------------------------
    # Overall percentage of correct predictions
    accuracy = accuracy_score(y_test, y_pred)

    # ------------------------------------------
    # 2. Precision
    # ------------------------------------------
    # Of all customers predicted as churn,
    # how many actually churned?
    precision = precision_score(y_test, y_pred, zero_division=0)

    # ------------------------------------------
    # 3. Recall
    # ------------------------------------------
    # Of all customers who actually churned,
    # how many were correctly identified?
    recall = recall_score(y_test, y_pred, zero_division=0)

    # ------------------------------------------
    # 4. Confusion Matrix
    # ------------------------------------------
    # [[TN, FP],
    #  [FN, TP]]
    cm = confusion_matrix(y_test, y_pred)

    # Save all metrics including confusion matrix
    results.append({
        "Model": model_name,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "Confusion Matrix": cm.tolist()
    })

    # Print individual model results
    print("\n==============================")
    print("Model:", model_name)
    print("==============================")
    print(f"Accuracy : {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall   : {recall:.4f}")
    print("Confusion Matrix:")
    print(cm)

# Convert results into a DataFrame
results_df = pd.DataFrame(results)

# Sort models by highest accuracy
results_df = results_df.sort_values(
    by="Accuracy",
    ascending=False
).reset_index(drop=True)

# Display final comparison table
print("\n==============================")
print("MODEL COMPARISON")
print("==============================")
print(results_df)
```
